# Clasificacion

Conjunto de datos gaussianos para tareas de clustering. Se puede seleccionar un numero total de muestras con **n_samples**, numero de caracteristicas con **n_features** y el numero de objetivos.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# seleccion de 200 muestras
X,y = make_blobs(n_samples=200, n_features=2, centers=3, random_state=10)
```

```python
plt.figure(figsize=(4,4))
# grafico con representacion de 3 clases con distintas figuras de representacion
for target, color, marker in zip(range(3),['r','g','b'],['o','^','+']):
    plt.scatter(X[y==target,0],X[y==target,1],c=color,marker=marker,label='class'+format(target))
    
    plt.xlabel('Caracteristica  1', fontsize=15)
    plt.ylabel('Caracteristica  2', fontsize=15)
    plt.title('Tres Categorias')
    plt.legend(prop={'size':14}); plt.grid(True)
   
```
